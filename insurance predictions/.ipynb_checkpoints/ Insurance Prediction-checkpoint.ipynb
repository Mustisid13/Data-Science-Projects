```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OrdinalEncoder
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
train_data = pd.read_csv('TRAIN.csv')
```

```python
train_data.head()
```

```python
train_data.dtypes
```

```python
train_data = train_data.drop('id',axis=1)
```

```python
train_data.isnull().sum()
```

```python
categorical_columns = ['Gender','Vehicle_Age','Vehicle_Damage']
continous_columns = [col for col in train_data.columns if col not in categorical_columns]
continous_columns = continous_columns[:-1] # removing response column
```

```python
continous_columns
```

# Balancing Data

```python
# sepearating data based on target value
response_0 = train_data[train_data['Response']==0]
response_1 = train_data[train_data['Response']==1]
print(f'Response 0: {int((len(response_0)/len(train_data)) *100)}% \n Response 1: {int((len(response_1)/len(train_data)) *100)}%')
```

```python
import random
```

```python
# taking random values from response 0 equal to size of response 1

balanced = random.sample(response_0.values.tolist(),37300) 
balanced = pd.DataFrame(balanced)
balanced.columns = train_data.columns
balanced.head()
```

```python
[{"Gender":"Female", "Age":43, "Driving_License":1,"Region_Code":12,"Previously_Insured":0,"Vehicle_Age":">2Years","Vehicle_Damage":"yes",
      "Annual_Premium":34231,"Policy_Sales_Channel":32,"Vintage":54}]
```

```python
final_data = pd.concat([balanced,response_1],ignore_index=True) # concatinating both dataframes
```

```python
from sklearn.utils import shuffle
final_data = shuffle(final_data) # shuffling data
final_data.head()
```

## Encoding and Scaling Data

```python
from sklearn.compose import ColumnTransformer

data_transformer = ColumnTransformer([
    ('enc',OrdinalEncoder(),categorical_columns),
    ('scaler',MinMaxScaler(),continous_columns)
])
final_data_transformed = pd.DataFrame(data_transformer.fit_transform(final_data.drop(['Response'],axis=1)),columns=train_data.columns[0:10])
```

```python
final_data_transformed['Response'] = final_data['Response'].values
```

```python
final_data_transformed.head()
```

## Data Visualization

```python
from pandas.plotting import scatter_matrix
scatter_matrix(final_data_transformed[['Region_Code','Vehicle_Age','Annual_Premium','Policy_Sales_Channel']],figsize=(10,12))
```

## Loading and Preprocessing Test Data

```python
X = final_data_transformed.drop('Response',axis=1)
y = final_data_transformed['Response']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
test = pd.read_csv('TEST.csv')
test = test.drop(['id'],axis=1)
test.isnull().sum()
```

```python

test_transformed = data_transformer.transform(test)
test_transformed = pd.DataFrame(test_transformed,columns=test.columns)
```

```python
test_transformed.head()
```

## Training Model

```python

lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
# base model
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
```

```python
n_estimators = [200,300,350,400]
max_depth = [None,5,10,20]
max_features = [3,8,9,10]

# define grid search
rf = RandomForestClassifier()
grid = dict(n_estimators=n_estimators,max_features=max_features,max_depth=max_depth)

random_search = RandomizedSearchCV(estimator=rf, param_distributions=grid, n_jobs=-1, cv=6, scoring='accuracy',error_score=0)
random_result = random_search.fit(X, y)

# summarize results
print("Best: %f using %s" % (random_search.best_score_, random_search.best_params_))

means = random_search.cv_results_['mean_test_score']
stds = random_search.cv_results_['std_test_score']
params = random_search.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

# Saving Model and data transformer

```python
import joblib
joblib.dump(random_search.best_estimator_,'model.pkl')
joblib.dump(data_transformer,'data_transformer.pkl')
```
